```python
import yfinance as yf

def get_stock_price(ticker):
    stock = yf.Ticker(ticker)
    stock_info = stock.history(period="1d")
    current_price = stock_info['Close'].iloc[-1]  # Gets the latest close price
    return current_price
```

```python
import yfinance as yf
import pandas as pd

# Portfolio setup
initial_investment = 10000
allocations = {
    'BTC-USD': 0.125,
    'ETH': 0.075,
    'NVDA': 0.15,
    'AAPL': 0.12,
    'MSFT': 0.12,
    'TSLA': 0.11,
    'LLY': 0.10,
    'COIN': 0.05,
    'PLTR': 0.10,
    'IAG': 0.05
}
tickers = list(allocations.keys())
start_date = "2024-12-09"  # Adjust the start date as needed
end_date = "2025-01-28"    # Adjust the end date as needed

# Download historical prices for each ticker
data = yf.download(tickers, start=start_date, end=end_date)['Close']

# Calculate initial shares based on initial prices and allocations
initial_prices = data.iloc[0]
shares = {
    ticker: (initial_investment * allocation) / initial_prices[ticker]
    for ticker, allocation in allocations.items()
}

# Create a DataFrame to store daily portfolio values
daily_values = pd.DataFrame(index=data.index, columns=['Portfolio Value'])

# Calculate portfolio value for each day based on actual market prices
for date in data.index:
    portfolio_value = sum(data.loc[date, ticker] * shares[ticker] for ticker in tickers)
    daily_values.loc[date, 'Portfolio Value'] = portfolio_value

# Convert the index to just the date and reset it as a column
daily_values.index = pd.to_datetime(daily_values.index).date  # Ensure only date part
daily_values.reset_index(inplace=True)
daily_values.rename(columns={'index': 'Day'}, inplace=True)

# Display the resulting daily portfolio values DataFrame
daily_values
```

```python
forestfunds = daily_values
```

```python
# Calculate the Up/Down and Percentage Change columns
daily_values['Up/Down'] = daily_values['Portfolio Value'].diff().apply(lambda x: 'Up' if x > 0 else 'Down')
daily_values['Percentage Change'] = daily_values['Portfolio Value'].pct_change().apply(lambda x: f"{x * 100:.2f}%" if x >= 0 else f"({-x * 100:.2f}%)")

daily_values
```

```python
# Add a "Day of Week" column
daily_values.insert(1, 'Day of Week', pd.to_datetime(daily_values['Day']).dt.day_name())
daily_values
```
