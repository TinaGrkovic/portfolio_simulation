# Stock Prices

```python
# Download raw stock prices (without allocation)
data = yf.download(tickers, start=start_date, end=end_date)['Close']

# Ensure the index (date) is properly set and timezone naive
data.index = pd.to_datetime(data.index).tz_localize(None)

# Reset the index to move the date into a column
data.reset_index(inplace=True)

# Fill missing values with 0 for stocks that didn't exist in the period
#data.fillna(0, inplace=True)
data.dropna(inplace=True)


# Add an explicit index column
data.insert(0, 'Index', range(1, len(data) + 1))

# Save the raw data for Streamlit visualization
#data.to_csv('tina_individual_stock_prices.csv', index=False)
```

# Daily returns

```python
# Filter data to start from competition start date
competition_start_date = "2024-12-09"
competition_data = data[data['Date'] >= competition_start_date].copy()

# Calculate initial investment values for each stock based on percentage allocation
initial_values = {ticker: initial_investment * allocations[ticker] for ticker in tickers}

# Calculate the initial stock prices on the competition start date
initial_prices = competition_data.iloc[0][tickers]

# Calculate daily portfolio values based on price fluctuations
for stock in tickers:
    competition_data[stock] = (competition_data[stock] / initial_prices[stock]) * initial_values[stock]

# Sum up daily returns for total portfolio value change
#competition_data['Total Return'] = competition_data[tickers].sum(axis=1)

# Save daily cash flow changes
competition_data.reset_index(drop=True, inplace=True)
returns = competition_data

# Uncomment the line below to save the processed returns data to CSV
#competition_data.to_csv('tina_stock_daily_returns.csv', index=False)
```

# ALL the stock prices

```python
import yfinance as yf
import pandas as pd

# List of stock tickers from competition portfolios
portfolio_tickers = [
    "AAPL", "MSFT", "NVDA", "TGT", "LLY", "GOOG", "AMZN", "BTC-USD", "ETH-USD", "TSLA", 
    "BRK-B", "NTR", "FLUT", "DECK", "GOLD", "PATH", "ANET", "CTRA", "BORR", "RGTI",
    "PRKR", "TMDX", "ZETA", "FTCI", "LEAT", "LNTH", "TREE", "BLBD", "NXT", "QXO", "TSSI", 
    "UBER", "AVGO", "ELF", "MSCI", "SWPPX", "FXAIX", "BABA", "PFE", "TM", "UL", "JPM"
]

# Additional stocks from the provided image
additional_tickers = [
    "SPY",    # S&P 500
    "TSLA",   # Tesla
    "AAPL",   # Apple
    "KO",     # Coca-Cola
    "BTC-USD",# Bitcoin
    "^IXIC",   # Nasdaq
    "DIA",    # Dow Jones (Dow 30)
    "PFE",    # Pfizer
    "TM",     # Toyota
    "BABA",   # Alibaba
    "UL",     # Unilever
    "JPM"     # JPMorgan Chase
]

# Combine all tickers (ensuring no duplicates)
all_tickers = list(set(portfolio_tickers + additional_tickers))

# Define competition start date
start_date = "2024-12-09"
end_date = "2025-01-22"

# Download stock prices from Yahoo Finance
stock_data = yf.download(all_tickers, start=start_date, end=end_date)['Close']

# Handle missing values
# 1. Fill forward missing values for existing stocks
stock_data.fillna(method='ffill', inplace=True)

# 2. Fill backward to handle leading NaNs
stock_data.fillna(method='bfill', inplace=True)

# 3. Drop any remaining NaN values if still present
stock_data.dropna(axis=0, how='any', inplace=True)

# Reset index to store date column properly
stock_data.reset_index(inplace=True)

# Save the cleaned data to a CSV file
output_file = "all_stock_prices_since_competition_start.csv"
stock_data.to_csv(output_file, index=False)

print(f"Stock data successfully saved to {output_file}")
```
